## 3. Montagem da tabela geral

```python
# Criando as colunas na tabela geral // Atendimento

atendimento_columns = list(atendimento.columns).index('CD_PESSOA_FISICA')

atendimento.insert(atendimento_columns + 1, 'NM_PESSOA_PESQUISA', '')
atendimento.insert(atendimento_columns + 2, 'DT_NASCIMENTO', '')
atendimento.insert(atendimento_columns + 3, 'IE_SEXO', '')
```

```python
# Loop for para preencher as colunas de acordo com a tabela pessoa_fisica_pac

for index, row in atendimento.iterrows():
    if row['CD_PESSOA_FISICA'] in pessoa_fisica_pac['CD_PESSOA_FISICA'].values:
        match = pessoa_fisica_pac[
            pessoa_fisica_pac['CD_PESSOA_FISICA'] == row['CD_PESSOA_FISICA']]

        dt_nascimento = match.iloc[0]['DT_NASCIMENTO']
        ie_sexo = match.iloc[0]['IE_SEXO']

        atendimento.at[index, 'DT_NASCIMENTO'] = dt_nascimento
        atendimento.at[index, 'IE_SEXO'] = ie_sexo
```
